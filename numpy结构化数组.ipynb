```python
#以二进制的方式存储结构化的数据
records=[(9,5.3,4.5,5.8),(6,7.8,9.0,8.1),(12,13.4,56.7,34.2)]
fmt='<i3d'
def write_records(records,format,f):
    records_struct=struct.Struct(format)
    for r in records:
        f.write(records_struct.pack(*r))    
with open('rdata','wb') as f:
    write_records(records,fmt,f)   
```

```python
#以结构化数组的方式读取文件数据
f=open('rdata','rb')
trec=np.fromfile(f,dtype='<i,<d,<d,<d')  
f.close()
trec
```

```python
#https://www.pythonheidong.com/blog/article/167044/
tdxfields=['trade_date','open','high','low','close','amount','vol','unused']
fieldformats=['<i4','<i4','<i4','<i4','<i4','f4','<i4','<i4']
fieldtypes = np.dtype({'names': tdxfields, 'formats': fieldformats})
with open('sh601138.day', 'rb') as f:
    tdata = np.fromfile(f,dtype=fieldtypes)
```
