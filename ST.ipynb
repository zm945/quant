```python
Stock_Data_Dir=Path.home()/'Stock_Data/Tushare'
if (not Stock_Data_Dir.exists()):
    Stock_Data_Dir.mkdir()
out_csv_dir=Stock_Data_Dir/'csv'
if(not out_csv_dir.exists()):  #Path.exists()
    out_csv_dir.mkdir()
```

```python
tokenfile=Stock_Data_Dir/'Token'/'token.pk'
with open(tokenfile,'rb') as f :
    mytoken=pickle.load(f)
ts.set_token(mytoken)   
pro = ts.pro_api()
```

```python
AStock= pro.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,market')
AStock.market.value_counts()  
```

```python
clist=[]
for code in AStock.ts_code:
    df = pro.namechange(ts_code=code, fields='ts_code,name,start_date,end_date,change_reason')
    clist.append(df)
namehistory= pd.concat(clist,ignore_index=True)   
```

```python
ST=namehistory[namehistory.change_reason=='ST']
```
